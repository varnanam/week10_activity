### 4. Carry out correlation analysis on the dataset and find the correlation factor   between                                                  a. ph and Chloramines                                                                                                                                                              b. ph and Conductivity.                                                                                                                                                  What do we infer from this correlation value?

```python
data1=data.corr()
data1
```

```python
plt.subplots(figsize=(20,8))
sns.heatmap(data1,linewidth=.2,annot=True,vmin=-.2,vmax=1,cmap='YlGnBu')
plt.show()
```

```python
x=data[['ph','Chloramines','Conductivity']]
cor_matrix=x.corr()
cor_matrix
```

```python
plt.subplots(figsize=(8,4))
sns.heatmap(cor_matrix,linewidth=.2,annot=True,vmin=-.05,vmax=1,cmap='RdBu')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.pairplot(data,x_vars=['Chloramines','Conductivity'],y_vars='ph',kind='reg')
plt.tight_layout()
plt.show()
```

###   a. ph and Chloramines

 from the above data the correlation value for ph and chloramines is -0.025 which is closer to zero, so there is no correlation between these two features

###   b. ph and Conductivity
